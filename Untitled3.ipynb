Overall crime rate of Chicago,eespecially homocides, is substantially higher than the US average.
This analysis exlpores the crime rate from year 2012 to year 2017and compare the number of arrests with the number of crimes.


```python
import pandas as pd
from pandas import read_csv
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.offline as py
import seaborn as sns

```

```python
crimes = read_csv('Chicago_Crimes_2012_to_2017.csv', index_col='Date')
```

```python
crimes.head()
```

```python
crimes.info()
```

```python
crimes.index = pd.to_datetime(crimes.index)
```

```python
crimes.head()
```

Explore number of crimes throughout the years 2012 to 2016.As we can see the highest number of murders takes place between 2012 and 2013 and the lowest is in 2016, even though the media claimed that 2016 was the year with the highest crime rate in the city of Chicago.

```python
plt.figure(figsize=(20,10))

crimes.resample('M').size().plot(legend=False)
plt.title('Number of crimes per year (2012 - 2016)')
plt.xlabel('Years')
plt.ylabel('Number of crimes')
plt.show()
```

The graph below demonstrates the Types of crime.

```python
plt.figure(figsize=(20,10))

crimes.groupby([crimes['Primary Type']]).size().sort_values(ascending=True).plot(kind='barh')
plt.title('Number of crimes by type')
plt.ylabel('Crime Type')
plt.xlabel('Number of crimes')
plt.show()
```

```python
The graph below demonstates the most usual places where a crime takes place.
```

```python
plt.figure(figsize=(20,30))

crimes.groupby([crimes['Location Description']]).size().sort_values(ascending=True).plot(kind='barh')
plt.title('Number of crimes by Location')
plt.ylabel('Crime Location')
plt.xlabel('Number of crimes')
plt.show()
```

```python
domestic_yearly = crimes[crimes['Domestic'] == True]['Domestic']
print(domestic_yearly.head())
```

The most exceptional is the rate of domestic violence which is extremely lower in 2016 than than the previous years.

```python
domestic_yearly.resample('A').sum().plot()
plt.title('Yearly domestic violence')
plt.show()
```

```python
crimes1 = pd.DataFrame(crimes[crimes['Primary Type'].isin(['THEFT','BATTERY', 'CRIMINAL DAMAGE', 'NARCOTICS', 'ASSAULT', 'PROSTITUTION'])]['Primary Type'])
```

```python
crimes2 = crimes1.groupby([pd.TimeGrouper('M'), 'Primary Type'])
```

```python
data = crimes2['Primary Type'].count().unstack()
```

```python
data.plot()
plt.title("Top 5 monthly crimes")
plt.show()
```
